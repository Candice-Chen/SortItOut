```python
#differences between pdb and em map
import seaborn as sns
sns.set(style="whitegrid")
import pandas as pd
diff_p = pd.read_csv("diff_plot.csv")
sns.set_palette("Set3")
ax = sns.violinplot(x=diff_p["group"],y=diff_p["diff pdb em"],hue=diff_p["Linreg"],split=True,inner="quartile") 
```

```python
#pdb volume spread before and after linear regression
sns.set(style="whitegrid")
pdb_p = pd.read_csv("pdbvol_linreg.csv")
sns.set_palette("Set3")
ax = sns.violinplot(x=pdb_p["GROUP"],y=pdb_p["PDBVOL"],hue=pdb_p["LINREG"],split=True,inner="quartile") 

```

```python
#before and after removing group b structures with missing density
sensitivity=[0.45,0.65,0.7,0.75,0.6,0.733333333,0.8,0.8]
specificity=[0.865789474,0.823684211,0.789473684,0.702631579,0.805194805,0.733766234,0.694805195,0.564935065]
removed=['before','before','before','before','after','after','after','after']
clusters=['nn_3','nn_4','sw_2','sw_3','nn_3','nn_4','sw_2','sw_3']

df= pd.DataFrame({'removed':removed,'clusters':clusters, 'sensitivity':sensitivity, 'specificity':specificity})
df
```

```python
ax=sns.relplot(data=df,s=100, x="sensitivity", y="specificity", hue= "removed", style="clusters")
sns.set(style="darkgrid")
sns.set_palette("Set3")
ax.set(ylim=(0,1))
ax.set(xlim=(0,1))
```
